```python
golf = pd.read_csv('/content/drive/MyDrive/DTgolf.csv')
golf
```

```python
X_train = golf.drop('playgolf', axis=1).values
Y_train = golf['playgolf'].values

model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image

dot_data = export_graphviz(model, out_file=None, feature_names=['outlook','temp','humidity','windy'], class_names=[ '0' ,'1'], filled=True,  impurity=True, rounded=True)
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
